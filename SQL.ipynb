## Q1 --- List all the directors who directed a 'Comedy' movie in a leap year. (You need to check that the genre is 'Comedyâ€™ and year is a leap year) Your query should return director name, the movie name, and the year.

<h4>To determine whether a year is a leap year, follow these steps:</h4>

<ul>
    <li><b>STEP-1:</b> If the year is evenly divisible by 4, go to step 2. Otherwise, go to step 5.</li>
    <li><b>STEP-2:</b> If the year is evenly divisible by 100, go to step 3. Otherwise, go to step 4.</li>
    <li><b>STEP-3:</b> If the year is evenly divisible by 400, go to step 4. Otherwise, go to step 5.</li>
    <li><b>STEP-4:</b> The year is a leap year (it has 366 days).</li>
    <li><b>STEP-5:</b> The year is not a leap year (it has 365 days).</li>
</ul>

Year 1900 is divisible by 4 and 100 but it is not divisible by 400, so it is not a leap year.

```python
%%time
def grader_1(q1):
    q1_results  = pd.read_sql_query(q1,conn)
    print(q1_results.head(10))
    print(q1_results.shape)
    assert (q1_results.shape == (232,3))

query1 = """ SELECT p.Name Director_name,m.title Movie,m.year yea \
                          FROM Movie m , M_Director m_d,Genre g,M_Genre m_g,Person p \
                          ON g.Name LIKE '%Comedy%' AND TRIM(m_d.PID)=TRIM(p.PID) AND TRIM(g.GID) = TRIM(m_g.GID) AND TRIM(m_g.MID)=TRIM(m.MID) AND TRIM(m.MID)=TRIM(m_d.MID) \
                          AND (yea % 4 = 0)  AND ((yea % 100 <> 0) OR (yea % 400 = 0)) ORDER BY m.year """
grader_1(query1)
```

```python
%%time
def grader_1(q1):
    q1_results  = pd.read_sql_query(q1,conn)
    print(q1_results.head(10))
    print(q1_results.shape)
    assert (q1_results.shape == (232,3))

query1 = """ SELECT p.Name Director_name,m.title Movie,m.year \
                          FROM Movie m INNER JOIN M_Director m_d ON TRIM(m.MID) = TRIM(m_d.MID) \
                          INNER JOIN person p on TRIM(p.PID) = TRIM(m_d.PID) \
                          INNER JOIN M_Genre m_g ON TRIM(m_g.MID) = TRIM(m.MID)\
                          INNER JOIN genre g ON TRIM(m_g.GID) = TRIM(g.GID)\
                          WHERE g.name = '%Comedy%' AND (m.year % 4 = 0)  AND (m.year % 100 <> 0 OR m.year % 400 = 0) """
grader_1(query1)
```

```python
%%time

def grader_3(q3):
    q3_results  = pd.read_sql_query(q3,conn)
    print(q3_results.head(10))
    print(q3_results.shape)
    assert (q3_results.shape == (300,1))

query3 = """ WITH
    ACTORS_BEFORE_1970 AS
    (
        SELECT
            DISTINCT
            TRIM(MC.PID) PID
        FROM
                    Movie M
            JOIN    M_Cast MC
            ON      M.MID = MC.MID
        WHERE
            CAST(SUBSTR(M.year,-4) AS UNSIGNED) < 1970
    ),
    ACTORS_AFTER_1990 AS
    (
        SELECT
            DISTINCT
            TRIM(MC.PID) PID
        FROM
                    Movie M
            JOIN    M_Cast MC
            ON      M.MID = MC.MID
        WHERE
            CAST(SUBSTR(M.year,-4) AS UNSIGNED) > 1990
    )
    SELECT
        DISTINCT
        TRIM(P.Name) Actor_Name
    FROM
                ACTORS_BEFORE_1970 A_1970
        JOIN    ACTORS_AFTER_1990 A_1990
        ON      A_1970.PID = A_1990.PID
        JOIN    Person P
        ON      A_1970.PID = TRIM(P.PID) """

grader_3(query3)
```

## Q4 --- List all directors who directed 10 movies or more, in descending order of the number of movies they directed. Return the directors' names and the number of movies each of them directed.

```python
%%time
def grader_4(q4):
    q4_results  = pd.read_sql_query(q4,conn)
    print(q4_results.head(10))
    print(q4_results.shape)
    assert (q4_results.shape == (58,2))

query4 = """SELECT DISTINCT p.Name Director,COUNT(*) number_of_movies \
                        FROM Person p, M_Director m_d \
                        ON TRIM(p.PID) = TRIM(m_d.PID) \
                        GROUP BY TRIM(m_d.PID) HAVING COUNT(*) >=10 ORDER BY number_of_movies DESC """
grader_4(query4)
```

## Q8 --- Find all the actors that made more movies with Yash Chopra than any other director.

```python
%%time

def grader_8(q8):
    q8_results  = pd.read_sql_query(q8,conn)
    print(q8_results.head(10))
    print(q8_results.shape)
    assert (q8_results.shape == (245, 2))

query8 = """ select p.name,h.count 
from(select TRIM(mc.pid) mcpid,TRIM(md.pid) mdpid,count(TRIM(mc.MID)) count   
from m_cast mc
                 join m_director md 
                     on TRIM(md.MID)=TRIM(mc.MID)               
                 group by TRIM(mc.pid) ,TRIM(md.pid) 
                ) h
            join person p 
                on h.mcpid=TRIM(p.pid)
            where h.count = (select count(*) count   
                             from m_cast mc
                             join m_director md 
                                 on TRIM(md.mid)=TRIM(mc.mid)
                             where TRIM(mc.pid)=h.mcpid 
                             group by TRIM(mc.pid),TRIM(md.pid) 
                             order by count(*) desc
                             limit 1)
            and h.mdpid = (select pid 
                           from person 
                           where name like '%Yash Chopra%'
                          )
            order by h.count desc """

grader_8(query8)
```
